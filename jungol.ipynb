#3333: 막대기
image.png

image.png

```python
# 입력 처리

n = int(input())
h = [0]*n

for i in range(n):
    h[i] = int(input())

# 뒷 막대기보다 높은 막대기 갯수 세기
max = h[-1]
v = 1
for i in range(n-2, -1, -1):
    if h[i] > max:
        max = h[i]
        v = v + 1
print(v)
```

#827 : 반복제어문 1 - 형성평가 1
image.png

```python
# for 반복문을 이용하는 방법

n = int(input())
for i in range(n):
    print(i+1, end=" ")
```

```python
# while 반복문을 이용하는 방법

n = int(input())
i = 1
while i <= n:
    print(i, end=" ")
    i = i + 1
```
